```python
def add_disease_to_graph(tx, name, symptoms, causes, treatments):
    query = (
        "MERGE (d:Disease {name: $name}) "
        "WITH d "
        "UNWIND $symptoms AS symptom "
        "MERGE (s:Symptom {name: symptom}) "
        "MERGE (d)-[:HAS_SYMPTOM]->(s) "
        "WITH d "
        "UNWIND $causes AS cause "
        "MERGE (c:Cause {name: cause}) "
        "MERGE (d)-[:HAS_CAUSE]->(c) "
        "WITH d "
        "UNWIND $treatments AS treatment "
        "MERGE (t:Treatment {name: treatment}) "
        "MERGE (d)-[:HAS_TREATMENT]->(t)"
    )
    tx.run(query, name=name, symptoms=symptoms.split(', '), causes=causes.split(', '), treatments=treatments.split(', '))
```

```python
with driver.session() as session:
    for index, row in dfc_clean.iterrows():
        session.write_transaction(add_disease_to_graph, row['병명'], row['증상'], row['원인'], row['치료'])
```

## 증상 명사로 변경

```python
from konlpy.tag import Okt

# Okt 객체 생성
okt = Okt()
```

```python
stopwords = ['이', '있', '하', '것', '들', '그', '되', '수', '보', '않', '없', '나', '사람', '주', '아니', '등', '같', '우리', '때', '년', '가', '한', '지', '대하', '오', '말', '일', '그렇', '위하']
```

```python
# 증상 컬럼의 텍스트를 명사로 분석하고 불용어를 제거하는 함수
def extract_nouns_remove_stopwords(text):
    nouns = okt.nouns(text)
    nouns = [noun for noun in nouns if noun not in stopwords]
    return nouns

# 증상 컬럼에 함수 적용
dfc_clean['processed_증상'] = dfc_clean['증상'].apply(extract_nouns_remove_stopwords)

print(dfc_clean)
```

```python
dfc_clean['processed_원인'] = dfc_clean['원인'].apply(extract_nouns_remove_stopwords)
```

```python
dfc_clean['processed_치료'] = dfc_clean['치료'].apply(extract_nouns_remove_stopwords)
```
