```python
from sentence_transformers import SentenceTransformer
import pandas as pd
import numpy as np
import faiss
```

```python
embedding_function = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
dfc_clean['병명'] = dfc_clean['병명'].apply(lambda x: embedding_function.encode(x, convert_to_tensor=True))
```

```python
# 이전 단계는 동일하게 유지
embedding_function = SentenceTransformer("all-MiniLM-L6-v2")
dfc_clean['병명_벡터'] = dfc_clean['병명'].apply(lambda x: embedding_function.encode(x, convert_to_numpy=True))

# Faiss 인덱스 생성
dimension = dfc_clean['병명_벡터'].iloc[0].shape[0]  # 벡터의 차원 확인
index = faiss.IndexFlatL2(dimension)  # L2 거리를 사용하는 평면 인덱스

# 벡터를 Faiss 인덱스에 추가
for vec in dfc_clean['병명_벡터']:
    vec = np.array(vec).astype('float32').reshape(1, -1)  # 올바른 형태로 변환
    faiss.normalize_L2(vec)  # L2 정규화 적용
    index.add(vec)  # 인덱스에 벡터 추가
```
