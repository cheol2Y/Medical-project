```python
import psycopg2
import csv


# PostgreSQL에 연결 (비밀번호가 없는 경우)
conn = psycopg2.connect(
    host="localhost",
    database="test1",
    user="postgres"
)

# 커서 생성
cur = conn.cursor()

# 테이블 생성 쿼리
create_table_query = '''
CREATE TABLE IF NOT EXISTS my_table (
    doc_id SERIAL PRIMARY KEY,
    title VARCHAR(255),
    question TEXT,
    answer TEXT
);
'''

# 테이블 생성
cur.execute(create_table_query)

# CSV 파일에서 데이터 읽어서 테이블에 삽입
with open('/Users/jangminsoo/Desktop/dev/pro/Medical-project/csv_folder/today_naver_QnA/details.csv', 'r') as f:
    reader = csv.reader(f)
    next(reader)  # 첫 행은 헤더이므로 건너뜀
    for row in reader:
        cur.execute(
            "INSERT INTO my_table (title, question, answer) VALUES (%s, %s, %s)",
            (row[0], row[1], row[2])
        )

# 변경사항 커밋
conn.commit()

# 커서와 연결 닫기
cur.close()
conn.close()
```

```python
import psycopg2

# PostgreSQL에 연결
conn = psycopg2.connect(
    host="localhost",
    database="test1",
    user="postgres"
)

# 커서 생성
cur = conn.cursor()

# 테이블 목록 조회
cur.execute("SELECT table_name FROM information_schema.tables WHERE table_schema = 'public'")
tables = cur.fetchall()
print("테이블 목록:")
for table in tables:
    print(table[0])

# 데이터 조회
cur.execute("SELECT * FROM my_table")
rows = cur.fetchall()
print("\n테이블 데이터:")
for row in rows:
    print(row)

# 변경사항 커밋
conn.commit()

# 커서와 연결 닫기
cur.close()
conn.close()
```
