```python
# doctornow 질병백과 id 추출
from bs4 import BeautifulSoup
import requests

url = "https://doctornow.co.kr/content/encyclopedia"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
link_tags = soup.find_all('a', href=True)

# URL에서 ID 값을 추출하여 리스트에 저장
id_values = []
for link in link_tags:
    href = link['href']
    if "/content/encyclopedia/diseases/" in href:
        id_value = href.split('/')[-1]
        id_values.append(id_value)


id_values = list(set(id_values))
id_values.__len__()
```

```python
# 증상을 제대로 가져오는지 확인하기
import requests
from bs4 import BeautifulSoup

for id_value in id_values:
    # URL 생성
    url = f'https://doctornow.co.kr/content/encyclopedia/diseases/{id_value}'
    response = requests.get(url)
    
    # 요청이 성공했을 경우
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        # 증상 가져오기
        symptoms = soup.select('div.sc-73e569df-3.kWweaE')
        symptoms2 = soup.select('div.sc-e1293ba-3.jCrLxL')
        
        # 증상 출력
        print(f"증상 ({id_value}):")
        for symptom in symptoms:
            print(symptom.text.strip())
        for symptom in symptoms2:
            print(symptom.text.strip())
        print()
    else:
        print(f"URL 요청 실패: {url}")
```

```python
# 위에 코드에서 병명을 추가하고 치료 정보도 추가해서 가져오기
import requests
from bs4 import BeautifulSoup

for id_value in id_values:
    # URL 생성
    url = f'https://doctornow.co.kr/content/encyclopedia/diseases/{id_value}'
    response = requests.get(url)
    
    # 요청이 성공했을 경우
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        # 병명 가져오기
        disease_name_element = soup.find('h1', class_='sc-6dffaabc-1 gVbXOB')
        disease_name = disease_name_element.text.strip() if disease_name_element else '병명을 찾을 수 없음'
        
        # 증상 가져오기
        symptoms = soup.select('div.sc-73e569df-3.kWweaE')
        symptoms2 = soup.select('div.sc-e1293ba-3.jCrLxL')
        
        # 치료 가져오기
        treatment_element = soup.find('div', class_='sc-caf1a99f-3 fFaQVu')
        treatment = treatment_element.text.strip() if treatment_element else '치료 정보를 찾을 수 없음'
        
        # 출력
        print(f"병명: {disease_name}")
        print(f"증상 ({id_value}):")
        for symptom in symptoms:
            print(symptom.text.strip())
        for symptom in symptoms2:
            print(symptom.text.strip())
        print(f"치료: {treatment}\n")
    else:
        print(f"URL 요청 실패: {url}")
```

```python
# 개요 정보를 추가함 ( 의사 답변에서 병명을 뽑아내야 하는데 개요 정보가 가장 의사 답변에 근접하고 필요한 내용이 전부 들어 있다고 판단)
import os
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm

folder_name = 'doctornow_diease_wiki_plus_all_info'

if not os.path.exists(folder_name):
    os.makedirs(folder_name)

data = []

for id_value in tqdm(id_values, desc="진행 상황"):
    url = f'https://doctornow.co.kr/content/encyclopedia/diseases/{id_value}'
    response = requests.get(url)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # 개요 가져오기
        all_info_element = soup.find('div', class_='sc-b7599da3-4 fouhrE')
        all_info = all_info_element.text.strip() if all_info_element else '개요를 찾을 수 없음'
        
        # 병명 가져오기
        disease_name_element = soup.find('h1', class_='sc-6dffaabc-1 gVbXOB')
        disease_name = disease_name_element.text.strip() if disease_name_element else '병명을 찾을 수 없음'
        
        # 증상 가져오기
        symptoms = soup.select('div.sc-73e569df-3.kWweaE')
        symptoms2 = soup.select('div.sc-e1293ba-3.jCrLxL')
        
        # 치료 정보 가져오기
        treatment_element = soup.find('div', class_='sc-caf1a99f-3 fFaQVu')
        treatment = treatment_element.text.strip() if treatment_element else '치료 정보를 찾을 수 없음'
        
        data.append({
            '개요': all_info, 
            '병명': disease_name,
            '증상': [symptom.text.strip() for symptom in symptoms + symptoms2],
            '치료': treatment
        })
    else:
        print(f"URL 요청 실패: {url}")

df = pd.DataFrame(data)

# 각각의 병명에 해당하는 CSV 파일로 저장
for index, row in df.iterrows():
    # 파일 이름에 공백 제거 및 특수 문자 제거
    file_name = f"{folder_name}/{row['병명'].replace(' ', '').replace('/', '_')}.csv"
    row_df = pd.DataFrame({'개요': row['개요'], '병명' : row['병명'],'증상': row['증상'], '치료': row['치료']})
    row_df.to_csv(file_name, index=False)

print("모든 데이터를 저장했습니다.")
```

```python
import os
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# 한글 불용어 목록
stopwords = [
    '그', '저', '이', '저희', '요', '하다', '있다', '없다', '없이', '없는', '없는데', '있는데', '있으며', '있으나', '있지만',
    '않는', '않습니다', '않다', '않았다', '않았습니다', '되다', '되었습니다', '되고', '되고있는', '되어', '되어있으며', '되어있지만', '한다',
    '되어있다', '되어있는', '되었어요', '되었지만', '하고', '하지', '하고있는', '하고자', '하려고', '하기', '하기는', '하기로','때문에','것은', '인한',
    '하기를', '하기에는', '하기에도', '하기에', '하기만', '하기만하면', '하기도', '하기보다는', '하기보다', '하기보단','등의','때나',
    '하기보다도', '하기보담', '하지만', '그러나', '그러나무엇보다', '그런데', '그렇지만', '이거', '그거', '저거', '그냥',
    '이냥', '저냥', '이것', '그것', '저것', '이건', '그건', '저건', '이렇게', '저렇게', '그렇게', '이런', '그런',
    '저런', '이렇나', '그렇나', '저렇나', '이렇다', '그렇다', '저렇다', '그래도', '그렇지도', '이러면', '그러면', '이러니',
    '그러니', '저러니', '이러니저러니', '그러니그러니', '이러니', '그러니', '저러니', '이러니저러니', '그러니그러니',
    '이러면서', '그러면서', '저러면서', '이러나', '그러나', '저러나', '이러면', '그러면', '저러면', '이렇게되면', '그렇게되면',
    '저렇게되면', '이렇게되면서', '그렇게되면서', '저렇게되면서', '이렇지', '그렇지', '저렇지', '이러한', '그러한', '저러한',
    '이와', '그와', '저와', '이와같이', '그와같이', '저와같이', '이와비교하여', '그와비교하여', '저와비교하여', '이러한',
    '그러한', '저러한', '이와', '그와', '저와', '여기', '여기서', '여기에', '여기서부터', '여기부터', '여기에서', '여기로',
    '여기까지', '여기저기', '이때', '그때', '저때', '이번', '그번', '저번', '이렇게되면', '그렇게되면', '저렇게되면',
    '이렇게되면서', '그렇게되면서', '저렇게되면서', '이렇게하면', '그렇게하면', '저렇게하면', '이렇게해서', '그렇게해서', '저렇게해서',
    '그래', '그런지', '그러나', '그렇지만', '그러나', '그래도', '그러나', '그래도', '그러나', '그런', '않다', '되어다', '되다', '하다', '어떻다', '이렇다', '이다', '어제', '매일', 
    '아', '휴', '아이구', '아이쿠', '아이고', '어', '나', '우리', '저희', '따라', '의해', '을', '를', '에', '의', '가', '으로', '로', '에게', '뿐이다', '의거하여', '근거하여', 
    '입각하여', '기준으로', '예하면', '예를 들면', '예를 들자면', '저', '소인', '소생', '저희', '지말고', '하지마', '하지마라', '다른', '물론', '또한', '그리고', '비길수 없다', 
    '해서는 안된다', '뿐만 아니라', '만이 아니다', '만은 아니다', '막론하고', '관계없이', '그치지 않다', '그러나', '그런데', '하지만', '든간에', '논하지 않다', '따지지 않다', '설사', '비록',
    '더라도', '아니면', '만 못하다', '하는 편이 낫다', '불문하고', '향하여', '향해서', '향하다', '쪽으로', '틈타', '이용하여', '타다', '오르다', '제외하고', '이 외에', '이 밖에', '하여야', 
    '비로소', '한다면 몰라도', '외에도', '이곳', '여기', '부터', '기점으로', '따라서', '할 생각이다', '하려고하다', '이리하여', '그리하여', '그렇게 함으로써', '하지만', '일때', '할때', 
    '앞에서', '중에서', '보는데서', '으로써', '로써', '까지', '해야한다', '일것이다', '반드시', '할줄알다', '할수있다', '할수있어', '임에 틀림없다', '한다면', '등', '등등', '제', '겨우', 
    '단지', '다만', '할뿐', '딩동', '댕그', '대해서', '대하여', '대하면', '훨씬', '얼마나', '얼마만큼', '얼마큼', '남짓', '여', '얼마간', '약간', '다소', '좀', '조금', '다수', '몇', 
    '얼마', '지만', '하물며', '또한', '그러나', '그렇지만', '하지만', '이외에도', '대해 말하자면', '뿐이다', '다음에', '반대로', '반대로 말하자면', '이와 반대로', '바꾸어서 말하면', 
    '바꾸어서 한다면', '만약', '그렇지않으면', '까악', '툭', '딱', '삐걱거리다', '보드득', '비걱거리다', '꽈당', '응당', '해야한다', '에 가서', '각', '각각', '여러분', '각종', '각자', 
    '제각기', '하도록하다', '와', '과', '그러므로', '그래서', '고로', '한 까닭에', '하기 때문에', '거니와', '이지만', '대하여', '관하여', '관한', '과연', '실로', '아니나다를가', 
    '생각한대로', '진짜로', '한적이있다', '하곤하였다', '하', '하하', '허허', '아하', '거바', '와', '오', '왜', '어째서', '무엇때문에', '어찌', '하겠는가', '무슨', '어디', '어느곳', 
    '더군다나', '하물며', '더욱이는', '어느때', '언제', '야', '이봐', '어이', '여보시오', '흐흐', '흥', '휴', '헉헉', '헐떡헐떡', '영차', '여차', '어기여차', '끙끙', '아야', '앗', 
    '아야', '콸콸', '졸졸', '좍좍', '뚝뚝', '주룩주룩', '솨', '우르르', '그래도', '또', '그리고', '바꾸어말하면', '바꾸어말하자면', '혹은', '혹시', '답다', '및', '그에 따르는', '때가 되어',
    '즉', '지든지', '설령', '가령', '하더라도', '할지라도', '일지라도', '지든지', '몇', '거의', '하마터면', '인젠', '이젠', '된바에야', '된이상', '만큼', '어찌됏든', '그위에', '게다가', 
    '점에서 보아', '비추어 보아', '고려하면', '하게될것이다', '일것이다', '비교적', '좀', '보다더', '비하면', '시키다', '하게하다', '할만하다', '의해서', '연이서', '이어서', '잇따라', '뒤따라', '뒤이어', '결국', '의지하여', '기대여', '통하여', '자마자', '더욱더', '불구하고', '얼마든지', '마음대로', '주저하지 않고', '곧', '즉시', '바로', '당장', '하자마자', '밖에 안된다', '하면된다', '그래', '그렇지', '요컨대', '다시 말하자면', '바꿔 말하면', '즉', '구체적으로', '말하자면', '시작하여', '시초에', '이상', '허', '헉', '허걱', '바와같이', '해도좋다', '해도된다', '게다가', '더구나', '하물며', '와르르', '팍', '퍽', '펄렁', '동안', '이래', '하고있었다', '이었다', '에서', '로부터', '까지', '예하면', '했어요', '해요', '함께', '같이', '더불어', '마저', '마저도', '양자', '모두', '습니다', '가까스로', '하려고하다', '즈음하여', '다른', '다른 방면으로', '해봐요', '습니까', '했어요', '말할것도 없고', '무릎쓰고', '개의치않고', '하는것만 못하다', '하는것이 낫다', '매', '매번', '들', '모', '어느것', '어느', '로써', '갖고말하자면', '어디', '어느쪽', '어느것', '어느해', '어느 년도', '라 해도', '언젠가', '어떤것', '어느것', '저기', '저쪽', '저것', '그때', '그럼', '그러면', '요만한걸', '그래', '그때', '저것만큼', '그저', '이르기까지', '할 줄 안다', '할 힘이 있다', '너', '너희', '당신', '어찌', '설마', '차라리', '할지언정', '할지라도', '할망정', '할지언정', '구토하다', '게우다', '토하다', '메쓰겁다', '옆사람', '퉤', '쳇', '의거하여', '근거하여', '의해', '따라', '힘입어', '그', '다음', '버금', '두번째로', '기타', '첫번째로', '나머지는', '그중에서', '견지에서', '형식으로 쓰여', '입장에서', '위해서', '단지', '의해되다', '하도록시키다', '뿐만아니라', '반대로', '전후', '전자', '앞의것', '잠시', '잠깐', '하면서', '그렇지만', '다음에', '그러한즉', '그런즉', '남들', '아무거나', '어찌하든지', '같다', '비슷하다', '예컨대', '이럴정도로', '어떻게', '만약', '만일', '위에서 서술한바와같이', '인 듯하다', '하지 않는다면', '만약에', '무엇', '무슨', '어느', '어떤', '아래윗', '조차', '한데', '그럼에도 불구하고', '여전히', '심지어', '까지도', '조차도', '하지 않도록', '않기 위하여', '때', '시각', '무렵', '시간', '동안', '어때', '어떠한', '하여금', '네', '예', '우선', '누구', '누가 알겠는가', '아무도', '줄은모른다', '줄은 몰랏다', '하는 김에', '겸사겸사', '하는바', '그런 까닭에', '한 이유는', '그러니', '그러니까', '때문에', '그', '너희', '그들', '너희들', '타인', '것', '것들', '너', '위하여', '공동으로', '동시에', '하기 위하여', '어찌하여', '무엇때문에', '붕붕', '윙윙', '나', '우리', '엉엉', '휘익', '윙윙', '오호', '아하', '어쨋든', '만 못하다', '하기보다는', '차라리', '하는 편이 낫다', '흐흐', '놀라다', '상대적으로 말하자면', '마치', '아니라면', '쉿', '그렇지 않으면', '그렇지 않다면', '안 그러면', '아니었다면', '하든지', '아니면', '이라면', '좋아', '알았어', '하는것도', '그만이다', '어쩔수 없다', '하나', '일', '일반적으로', '일단', '한켠으로는', '오자마자', '이렇게되면', '이와같다면', '전부', '한마디', '한항목', '근거로', '하기에', '아울러', '하지 않도록', '않기 위해서', '이르기까지', '이 되다', '로 인하여', '까닭으로', '이유만으로', '이로 인하여', '그래서', '이 때문에', '그러므로', '그런 까닭에', '알 수 있다', '결론을 낼 수 있다', '으로 인하여', '있다', '어떤것', '관계가 있다', '관련이 있다', '연관되다', '어떤것들', '에 대해', '이리하여', '그리하여', '여부', '하기보다는', '하느니', '하면 할수록', '운운', '이러이러하다', '하구나', '하도다', '다시말하면', '다음으로', '에 있다', '에 달려 있다', '우리', '우리들', '오히려', '하기는한데', '어떻게', '어떻해', '어찌됏어', '어때', '어째서', '본대로', '자', '이', '이쪽', '여기', '이것', '이번', '이렇게말하자면', '이런', '이러한', '이와 같은', '요만큼', '요만한 것', '얼마 안 되는 것', '이만큼', '이 정도의', '이렇게 많은 것', '이와 같다', '이때', '이렇구나', '것과 같이', '끼익', '삐걱', '따위', '와 같은 사람들', '부류의 사람들', '왜냐하면', '중의하나', '오직', '오로지', '에 한하다', '하기만 하면', '도착하다', '까지 미치다', '도달하다', '정도에 이르다', '할 지경이다', '결과에 이르다', '관해서는', '여러분', '하고 있다', '한 후', '혼자', '자기', '자기집', '자신', '우에 종합한것과같이', '총적으로 보면', '총적으로 말하면', '총적으로', '대로 하다', '으로서', '참', '그만이다', '할 따름이다', '쿵', '탕탕', '쾅쾅', '둥둥', '봐', '봐라', '아이야', '아니', '와아', '응', '아이', '참나', '년', '월', '일', '령', '영', '일', '이', '삼', '사', '오', '육', '륙', '칠', '팔', '구', '이천육', '이천칠', '이천팔', '이천구', '하나', '둘', '셋', '넷', '다섯', '여섯', '일곱', '여덟', '아홉', '령', '이천', '이천일', '이천이', '이천삼', '이천사', '이천오', '이천육', '이천칠', '이천팔', '이천구', '삼천', '사천', '오천', '육천', '칠천', '팔천', '구천', '만', '영', '일', '이', '삼', '사', '오', '육', '륙', '칠', '팔', '구', '이천육', '이천칠', '이천팔', '이천구', '하나', '둘', '셋', '넷', '다섯', '여섯', '일곱', '여덟', '아홉', '령', '이천', '이천일', '이천이', '이천삼', '이천사', 
    '이천오', '이천육', '이천칠', '이천팔', '이천구', '삼천', '사천', '오천', '육천', '칠천', '팔천', '구천', '만' 
]

# 개요정보도 포함된 닥터나우 크롤링 (답변에서 병명 추출) 폴더 안의 모든 CSV 파일 경로 가져오기
directory = "doctornow_diease_wiki_plus_all_info"
csv_files = [file for file in os.listdir(directory) if file.endswith(".csv")]
documents = {}

for file in csv_files:
    file_path = os.path.join(directory, file)
    df = pd.read_csv(file_path)
    disease_name = os.path.splitext(file)[0]
    
    # 병명 열이 없으면 건너뜀
    if "병명" not in df.columns:
        continue
    
    all_info = df["개요"].dropna().str.cat(sep=' ')  # 개요 열 선택하여 문자열 결합
    symptoms = df["증상"].dropna().str.cat(sep=' ')  # 증상 열 선택하여 문자열 결합
    if not all_info and not symptoms:
        continue
    
    # 개요와 증상을 합쳐서 저장
    combined_text = " ".join([all_info, symptoms])
    documents[disease_name] = combined_text

# TF-IDF 벡터화 객체 생성
tfidf_vectorizer = TfidfVectorizer(stop_words=stopwords)

result_data = {}  # 최종 결과를 저장할 딕셔너리
for disease, text in documents.items():
    if not text:  # 문서가 비어 있는 경우 건너뜀
        continue
    tfidf_matrix = tfidf_vectorizer.fit_transform([text])
    feature_names = tfidf_vectorizer.get_feature_names_out()
    dense = tfidf_matrix.todense()
    disease_data = {}  # 각 병명에 대한 데이터를 저장할 딕셔너리
    for i in range(dense.shape[1]):
        keyword = feature_names[i]
        score = dense[0, i]
        disease_data[keyword] = score  # 키워드와 점수 저장
    result_data[disease] = disease_data  # 최종 딕셔너리에 저장

# 결과를 담을 리스트 생성
data = []

# 결과 확인을 위한 코드
for disease, keywords in result_data.items():
    # 키워드와 점수를 각각 리스트로 저장
    disease_list = list(disease)
    keyword_list = list(keywords.keys())
    score_list = list(keywords.values())
    
    # 데이터프레임으로 변환하여 리스트에 추가
    df = pd.DataFrame({'diease' : disease, 'keyword': keyword_list, 'score': score_list})
    data.append(df)

# 모든 데이터프레임을 합쳐서 하나의 데이터프레임으로 생성
final_df = pd.concat(data, ignore_index=True)

# 결과 확인
print(final_df)
```

```python
# 명사만 추출해서 중요도 계산하는 코드
import os
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from konlpy.tag import Okt

# KoNLPy의 형태소 분석기 객체 생성
okt = Okt()

# 한글 불용어 목록
stopwords = [
    '그', '저', '이', '저희', '요', '하다', '있다', '없다', '없이', '없는', '없는데', '있는데', '있으며', '있으나', '있지만',
    '않는', '않습니다', '않다', '않았다', '않았습니다', '되다', '되었습니다', '되고', '되고있는', '되어', '되어있으며', '되어있지만', '한다',
    '되어있다', '되어있는', '되었어요', '되었지만', '하고', '하지', '하고있는', '하고자', '하려고', '하기', '하기는', '하기로','때문에','것은', '인한',
    '하기를', '하기에는', '하기에도', '하기에', '하기만', '하기만하면', '하기도', '하기보다는', '하기보다', '하기보단','등의','때나',
    '하기보다도', '하기보담', '하지만', '그러나', '그러나무엇보다', '그런데', '그렇지만', '이거', '그거', '저거', '그냥',
    '이냥', '저냥', '이것', '그것', '저것', '이건', '그건', '저건', '이렇게', '저렇게', '그렇게', '이런', '그런',
    '저런', '이렇나', '그렇나', '저렇나', '이렇다', '그렇다', '저렇다', '그래도', '그렇지도', '이러면', '그러면', '이러니',
    '그러니', '저러니', '이러니저러니', '그러니그러니', '이러니', '그러니', '저러니', '이러니저러니', '그러니그러니',
    '이러면서', '그러면서', '저러면서', '이러나', '그러나', '저러나', '이러면', '그러면', '저러면', '이렇게되면', '그렇게되면',
    '저렇게되면', '이렇게되면서', '그렇게되면서', '저렇게되면서', '이렇지', '그렇지', '저렇지', '이러한', '그러한', '저러한',
    '이와', '그와', '저와', '이와같이', '그와같이', '저와같이', '이와비교하여', '그와비교하여', '저와비교하여', '이러한',
    '그러한', '저러한', '이와', '그와', '저와', '여기', '여기서', '여기에', '여기서부터', '여기부터', '여기에서', '여기로',
    '여기까지', '여기저기', '이때', '그때', '저때', '이번', '그번', '저번', '이렇게되면', '그렇게되면', '저렇게되면',
    '이렇게되면서', '그렇게되면서', '저렇게되면서', '이렇게하면', '그렇게하면', '저렇게하면', '이렇게해서', '그렇게해서', '저렇게해서',
    '그래', '그런지', '그러나', '그렇지만', '그러나', '그래도', '그러나', '그래도', '그러나', '그런', 
    '않다', '되어다', '되다', '하다', '어떻다', '이렇다', '이다', '어제', '매일', '아', '휴', '아이구', '아이쿠', '아이고', '어', '나', '우리', '저희', '따라', '의해', '을',
    '를', '에', '의', '가', '으로', '로', '에게', '뿐이다', '의거하여', '근거하여', '입각하여', '기준으로', '예하면', '예를 들면', '예를 들자면', '저', '소인', '소생', '저희', 
    '지말고', '하지마', '하지마라', '다른', '물론', '또한', '그리고', '비길수 없다', '해서는 안된다', '뿐만 아니라', '만이 아니다', '만은 아니다', '막론하고', '관계없이', '그치지 않다', '그러나', '그런데', '하지만', '든간에', '논하지 않다', '따지지 않다', '설사', '비록', '더라도', '아니면', '만 못하다', '하는 편이 낫다', '불문하고', '향하여', '향해서', '향하다', '쪽으로', '틈타', '이용하여', '타다', '오르다', '제외하고', '이 외에', '이 밖에', '하여야', '비로소', '한다면 몰라도', '외에도', '이곳', '여기', '부터', '기점으로', '따라서', '할 생각이다', '하려고하다', '이리하여', '그리하여', '그렇게 함으로써', '하지만', '일때', '할때', '앞에서', '중에서', '보는데서', '으로써', '로써', '까지', '해야한다', '일것이다', '반드시', '할줄알다', '할수있다', '할수있어', '임에 틀림없다', '한다면', '등', '등등', '제', '겨우', '단지', '다만', '할뿐', '딩동', '댕그', '대해서', '대하여', '대하면', '훨씬', '얼마나', '얼마만큼', '얼마큼', '남짓', '여', '얼마간', '약간', '다소', '좀', '조금', '다수', '몇', '얼마', '지만', '하물며', '또한', '그러나', '그렇지만', '하지만', '이외에도', '대해 말하자면', '뿐이다', '다음에', '반대로', '반대로 말하자면', '이와 반대로', '바꾸어서 말하면', '바꾸어서 한다면', '만약', '그렇지않으면', '까악', '툭', '딱', '삐걱거리다', '보드득', '비걱거리다', '꽈당', '응당', '해야한다', '에 가서', '각', '각각', '여러분', '각종', '각자', '제각기', '하도록하다', '와', '과', '그러므로', '그래서', '고로', '한 까닭에', '하기 때문에', '거니와', '이지만', '대하여', '관하여', '관한', '과연', '실로', '아니나다를가', '생각한대로', '진짜로', '한적이있다', '하곤하였다', '하', '하하', '허허', '아하', '거바', '와', '오', '왜', '어째서', '무엇때문에', '어찌', '하겠는가', '무슨', '어디', '어느곳', '더군다나', '하물며', '더욱이는', '어느때', '언제', '야', '이봐', '어이', '여보시오', '흐흐', '흥', '휴', '헉헉', '헐떡헐떡', '영차', '여차', '어기여차', '끙끙', '아야', '앗', '아야', '콸콸', '졸졸', '좍좍', '뚝뚝', '주룩주룩', '솨', '우르르', '그래도', '또', '그리고', '바꾸어말하면', '바꾸어말하자면', '혹은', '혹시', '답다', '및', '그에 따르는', '때가 되어', '즉', '지든지', '설령', '가령', '하더라도', '할지라도', '일지라도', '지든지', '몇', '거의', '하마터면', '인젠', '이젠', '된바에야', '된이상', '만큼', '어찌됏든', '그위에', '게다가', '점에서 보아', '비추어 보아', '고려하면', '하게될것이다', '일것이다', '비교적', '좀', '보다더', '비하면', '시키다', '하게하다', '할만하다', '의해서', '연이서', '이어서', '잇따라', '뒤따라', '뒤이어', '결국', '의지하여', '기대여', '통하여', '자마자', '더욱더', '불구하고', '얼마든지', '마음대로', '주저하지 않고', '곧', '즉시', '바로', '당장', '하자마자', '밖에 안된다', '하면된다', '그래', '그렇지', '요컨대', '다시 말하자면', '바꿔 말하면', '즉', '구체적으로', '말하자면', '시작하여', '시초에', '이상', '허', '헉', '허걱', '바와같이', '해도좋다', '해도된다', '게다가', '더구나', '하물며', '와르르', '팍', '퍽', '펄렁', '동안', '이래', '하고있었다', '이었다', '에서', '로부터', '까지', '예하면', '했어요', '해요', '함께', '같이', '더불어', '마저', '마저도', '양자', '모두', '습니다', '가까스로', '하려고하다', '즈음하여', '다른', '다른 방면으로', '해봐요', '습니까', '했어요', '말할것도 없고', '무릎쓰고', '개의치않고', '하는것만 못하다', '하는것이 낫다', '매', '매번', '들', '모', '어느것', '어느', '로써', '갖고말하자면', '어디', '어느쪽', '어느것', '어느해', '어느 년도', '라 해도', '언젠가', '어떤것', '어느것', '저기', '저쪽', '저것', '그때', '그럼', '그러면', '요만한걸', '그래', '그때', '저것만큼', '그저', '이르기까지', '할 줄 안다', '할 힘이 있다', '너', '너희', '당신', '어찌', '설마', '차라리', '할지언정', '할지라도', '할망정', '할지언정', '구토하다', '게우다', '토하다', '메쓰겁다', '옆사람', '퉤', '쳇', '의거하여', '근거하여', '의해', '따라', '힘입어', '그', '다음', '버금', '두번째로', '기타', '첫번째로', '나머지는', '그중에서', '견지에서', '형식으로 쓰여', '입장에서', '위해서', '단지', '의해되다', '하도록시키다', '뿐만아니라', '반대로', '전후', '전자', '앞의것', '잠시', '잠깐', '하면서', '그렇지만', '다음에', '그러한즉', '그런즉', '남들', '아무거나', '어찌하든지', '같다', '비슷하다', '예컨대', '이럴정도로', '어떻게', '만약', '만일', '위에서 서술한바와같이', '인 듯하다', '하지 않는다면', '만약에', '무엇', '무슨', '어느', '어떤', '아래윗', '조차', '한데', '그럼에도 불구하고', '여전히', '심지어', '까지도', '조차도', '하지 않도록', '않기 위하여', '때', '시각', '무렵', '시간', '동안', '어때', '어떠한', '하여금', '네', '예', '우선', '누구', '누가 알겠는가', '아무도', '줄은모른다', '줄은 몰랏다', '하는 김에', '겸사겸사', '하는바', '그런 까닭에', '한 이유는', '그러니', '그러니까', '때문에', '그', '너희', '그들', '너희들', '타인', '것', '것들', '너', '위하여', '공동으로', '동시에', '하기 위하여', '어찌하여', '무엇때문에', '붕붕', '윙윙', '나', '우리', '엉엉', '휘익', '윙윙', '오호', '아하', '어쨋든', '만 못하다', '하기보다는', '차라리', '하는 편이 낫다', '흐흐', '놀라다', '상대적으로 말하자면', '마치', '아니라면', '쉿', '그렇지 않으면', '그렇지 않다면', '안 그러면', '아니었다면', '하든지', '아니면', '이라면', '좋아', '알았어', '하는것도', '그만이다', '어쩔수 없다', '하나', '일', '일반적으로', '일단', '한켠으로는', '오자마자', '이렇게되면', '이와같다면', '전부', '한마디', '한항목', '근거로', '하기에', '아울러', '하지 않도록', '않기 위해서', '이르기까지', '이 되다', '로 인하여', '까닭으로', '이유만으로', '이로 인하여', '그래서', '이 때문에', '그러므로', '그런 까닭에', '알 수 있다', '결론을 낼 수 있다', '으로 인하여', '있다', '어떤것', '관계가 있다', '관련이 있다', '연관되다', '어떤것들', '에 대해', '이리하여', '그리하여', '여부', '하기보다는', '하느니', '하면 할수록', '운운', '이러이러하다', '하구나', '하도다', '다시말하면', '다음으로', '에 있다', '에 달려 있다', '우리', '우리들', '오히려', '하기는한데', '어떻게', '어떻해', '어찌됏어', '어때', '어째서', '본대로', '자', '이', '이쪽', '여기', '이것', '이번', '이렇게말하자면', '이런', '이러한', '이와 같은', '요만큼', '요만한 것', '얼마 안 되는 것', '이만큼', '이 정도의', '이렇게 많은 것', '이와 같다', '이때', '이렇구나', '것과 같이', '끼익', '삐걱', '따위', '와 같은 사람들', '부류의 사람들', '왜냐하면', '중의하나', '오직', '오로지', '에 한하다', '하기만 하면', '도착하다', '까지 미치다', '도달하다', '정도에 이르다', '할 지경이다', '결과에 이르다', '관해서는', '여러분', '하고 있다', '한 후', '혼자', '자기', '자기집', '자신', '우에 종합한것과같이', '총적으로 보면', '총적으로 말하면', '총적으로', '대로 하다', '으로서', '참', '그만이다', '할 따름이다', '쿵', '탕탕', '쾅쾅', '둥둥', '봐', '봐라', '아이야', '아니', '와아', '응', '아이', '참나', '년', '월', '일', '령', '영', '일', '이', '삼', '사', '오', '육', '륙', '칠', '팔', '구', '이천육', '이천칠', '이천팔', '이천구', '하나', '둘', '셋', '넷', '다섯', '여섯', '일곱', '여덟', '아홉', '령', '이천', '이천일', '이천이', '이천삼', '이천사', '이천오', '이천육', '이천칠', '이천팔', '이천구', '삼천', '사천', '오천', '육천', '칠천', '팔천', '구천', '만', '영', '일', '이', '삼', '사', '오', '육', '륙', '칠', '팔', '구', '이천육', '이천칠', '이천팔', '이천구', '하나', '둘', '셋', '넷', '다섯', '여섯', '일곱', '여덟', '아홉', '령', '이천', '이천일', '이천이', '이천삼', 
    '이천사', '이천오', '이천육', '이천칠', '이천팔', '이천구', '삼천', '사천', '오천', '육천', '칠천', '팔천', '구천', '만' 
]

# 개요정보도 포함된 닥터나우 크롤링 (답변에서 병명 추출) 폴더 안의 모든 CSV 파일 경로 가져오기
directory = "doctornow_diease_wiki_plus_all_info"
csv_files = [file for file in os.listdir(directory) if file.endswith(".csv")]
documents = {}

for file in csv_files:
    file_path = os.path.join(directory, file)
    df = pd.read_csv(file_path)
    disease_name = os.path.splitext(file)[0]
    
    # 병명 열이 없으면 건너뜀
    if "병명" not in df.columns:
        continue
    
    all_info = df["개요"].dropna().str.cat(sep=' ')  # 개요 열 선택하여 문자열 결합
    symptoms = df["증상"].dropna().str.cat(sep=' ')  # 증상 열 선택하여 문자열 결합
    if all_info == "" and symptoms == "":
        continue
    
    # 개요와 증상을 합쳐서 저장
    combined_text = f"{all_info} {symptoms}"
    
    # 명사만 추출하여 리스트에 저장
    nouns = okt.nouns(combined_text)
    
    # 불용어 제거
    clean_nouns = [word for word in nouns if word not in stopwords]
    
    # 리스트를 공백으로 구분된 문자열로 변환
    cleaned_text = " ".join(clean_nouns)
    
    documents[disease_name] = cleaned_text

# TF-IDF 벡터화 객체 생성
tfidf_vectorizer = TfidfVectorizer()

result_data = {}  # 최종 결과를 저장할 딕셔너리
for disease, text in documents.items():
    if not text:  # 문서가 비어 있는 경우 건너뜀
        continue
    tfidf_matrix = tfidf_vectorizer.fit_transform([text])
    feature_names = tfidf_vectorizer.get_feature_names_out()
    dense = tfidf_matrix.todense()
    disease_data = {}  # 각 병명에 대한 데이터를 저장할 딕셔너리
    for i in range(dense.shape[1]):
        keyword = feature_names[i]
        score = dense[0, i]
        disease_data[keyword] = score  # 키워드와 점수 저장
    result_data[disease] = disease_data  # 최종 딕셔너리에 저장

# 결과를 담을 리스트 생성
data = []

# 결과 확인을 위한 코드
for disease, keywords in result_data.items():
    # 키워드와 점수를 각각 리스트로 저장
    disease_list = [disease] * len(keywords)
    keyword_list = list(keywords.keys())
    score_list = list(keywords.values())
    
    # 데이터프레임으로 변환하여 리스트에 추가
    df = pd.DataFrame({'diease' : disease_list, 'keyword': keyword_list, 'score': score_list})
    data.append(df)

# 모든 데이터프레임을 합쳐서 하나의 데이터프레임으로 생성
final_df2 = pd.concat(data, ignore_index=True)

# 결과 확인
print(final_df2)
```

# 로지스틱 회귀 시험해보기

## 데이터 : 증상, 개요, 병명을 담은 닥터나우 질병백과


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.feature_extraction.text import TfidfVectorizer
import scipy.sparse as sp

# 주어진 데이터프레임을 사용하여 로지스틱 회귀 모델을 학습시키는 함수
def train_logistic_regression(final_df):
    # TF-IDF 벡터화 객체 생성
    tfidf_vectorizer = TfidfVectorizer()
    
    # TF-IDF 행렬 생성
    X_keyword_tfidf = tfidf_vectorizer.fit_transform(final_df['keyword'])
    
    # 'score' 열 추출
    X_score = final_df[['score']]
    
    # TF-IDF 행렬과 'score' 열을 수평으로 합침
    X_combined = sp.hstack((X_keyword_tfidf, sp.csr_matrix(X_score)), format='csr')
    
    # 특성과 타겟 분리
    y = final_df['diease']  # 병명을 타겟으로 사용
    
    # 학습 데이터와 테스트 데이터로 분할
    X_train, X_test, y_train, y_test = train_test_split(X_combined, y, test_size=0.2, random_state=42)
    
    # 로지스틱 회귀 모델 생성 및 학습
    logreg_model = LogisticRegression(max_iter=1000)  # 반복 횟수 조절 가능
    logreg_model.fit(X_train, y_train)
    
    # 테스트 데이터에 대한 예측
    y_pred = logreg_model.predict(X_test)
    
    # 정확도 평가
    accuracy = accuracy_score(y_test, y_pred)
    print("Accuracy:", accuracy)

# 주어진 데이터프레임을 사용하여 로지스틱 회귀 모델 학습
train_logistic_regression(final_df)
```
