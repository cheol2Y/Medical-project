```python
# 첫페이지에서 닥터나우 실시간 questionid값 가져오기
import requests
import json
from datetime import datetime, timedelta

# 현재 시간
now = datetime.now()

# 24시간 이전의 시간
twenty_four_hours_ago = now - timedelta(hours=24)

# 웹 페이지의 URL 설정
url = "https://doctornow.co.kr/content/qna"
response = requests.get(url)
html_text = response.text

# 문자열에서 JSON 데이터를 추출
json_data_start = html_text.find('{"props":')
json_data_end = html_text.find('</script>', json_data_start)
json_str = html_text[json_data_start:json_data_end]

# JSON 데이터 파싱
json_data = json.loads(json_str)

# 24시간 이내의 질문들만 추출
questions_within_24_hours = []

# 모든 질문에 대해 반복
for element in json_data['props']['pageProps']['dehydratedState']['queries'][0]['state']['data']['pages'][0]['newestQuestionCardCursorPage']['elements']:
    question_time = datetime.fromtimestamp(element['question']['createdAt'] / 1000)
    if question_time > twenty_four_hours_ago:
        questions_within_24_hours.append(element['question']['questionPid'])

# 결과 출력
print("24시간 이내의 질문의 questionPid:")
for question_pid in questions_within_24_hours:
    print(question_pid)
```
