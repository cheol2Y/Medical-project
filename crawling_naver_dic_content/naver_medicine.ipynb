```python
#의약품 id 리스트 생성
import requests
from bs4 import BeautifulSoup

page_count = 314
all_contents_ids = []  

for page_num in range(1, page_count + 1):
    url = f"https://terms.naver.com/list.naver?cid=67992&categoryId=67992&page={page_num}"
    response = requests.get(url)
    
    soup = BeautifulSoup(response.text, 'html.parser')

    contents_id_elements = soup.find_all(class_="__TLIST_PUT")

    # data-contentsid 값을 추출함!
    contents_ids = [element.get("data-contentsid") for element in contents_id_elements]
    all_contents_ids.extend(contents_ids)
    
print(all_contents_ids)
```

```python
# 내용을 잘 가져오는 테스트
url="https://terms.naver.com/entry.naver?docId=6555088&cid=67992&categoryId=67992"

from bs4 import BeautifulSoup
# HTML 데이터 (예제에서는 단순화된 버전 사용)
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
# 정보를 저장할 딕셔너리
info_dict = {}
# 각 섹션의 제목에 해당하는 h3 태그와 그에 따른 설명 p 태그를 찾아 딕셔너리에 저장
for h3_tag in soup.find_all("h3", class_="stress"):
    title = h3_tag.get_text().strip()
    content = h3_tag.find_next_sibling("p", class_="txt").get_text().strip()
    info_dict[title] = content
print(info_dict)
```

```python
import os
import pandas as pd
from bs4 import BeautifulSoup
import requests
from tqdm import tqdm

# 폴더 이름 설정
folder_name = "medicine"
if not os.path.exists(folder_name):
    os.makedirs(folder_name)

base_url = "https://terms.naver.com/entry.naver?docId="

# DataFrame을 저장할 빈 리스트 생성
dfs = []

# tqdm을 사용하여 진행상황 확인할 수 있게 함.
for content_id in tqdm(all_contents_ids, desc="진행 상황"):
    try:
        url = base_url + content_id

        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')

        # 의약품명
        medicine_name_tag = soup.find("h2", class_="headword")
        medicine_name = medicine_name_tag.text.strip().replace(" 개요정보", "") if medicine_name_tag else None

        # 진료과 가져오기
        profile_text = soup.find('div', class_='tmp_profile').get_text(strip=True)

        # 정보 딕셔너리 생성
        info_dict = {"의약품명": medicine_name, "업체명": profile_text}

        # 효능, 사용방법 등의 정보 가져오기
        for h3_tag in soup.find_all("h3", class_="stress"):
            title = h3_tag.get_text(strip=True)
            content = h3_tag.find_next_sibling("p", class_="txt").get_text(strip=True)

            # 키워드를 변경
            if title == "이 약의 효능은 무엇입니까?":
                title = "효능"
            elif title == "이 약은 어떻게 사용합니까?":
                title = "복용방법"
            elif title == "이 약의 사용상 주의사항은 무엇입니까?":
                title = "주의사항"
            elif title == "이 약은 어떤 이상반응이 나타날 수 있습니까?":
                title = "이상반응"
            elif title == "이 약은 어떻게 보관해야 합니까?":
                title = "보관방법"
            elif title == "이 약은 어떻게 보관해야 합니까?":
                title = "보관방법"
                

            info_dict[title] = content

        # 데이터프레임 생성
        df = pd.DataFrame.from_dict(info_dict, orient="index", columns=[content_id]).transpose()
        
        # 생성된 DataFrame을 리스트에 추가
        dfs.append(df)

    except Exception as e:
        print(f"Error occurred for content_id: {content_id}, {e}")

# 모든 DataFrame을 합쳐서 하나의 DataFrame으로 만듦
merged_df = pd.concat(dfs, ignore_index=True)

# 파일명 설정
file_name = "medicine.csv"

# 파일 경로 설정
file_path = os.path.join(folder_name, file_name)

# CSV 파일로 저장
merged_df.to_csv(file_path, encoding="utf-8-sig", index=False)

print("CSV 파일이 저장된 폴더:", folder_name)

```
