```python
from bs4 import BeautifulSoup
# HTML 데이터 (예제에서는 단순화된 버전 사용)
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
# 정보를 저장할 딕셔너리
info_dict = {}
# 각 섹션의 제목에 해당하는 h3 태그와 그에 따른 설명 p 태그를 찾아 딕셔너리에 저장
for h3_tag in soup.find_all("h3", class_="stress"):
    title = h3_tag.get_text().strip()
    content = h3_tag.find_next_sibling("p", class_="txt").get_text().strip()
    info_dict[title] = content
print(info_dict)
```

```python
# 지식백과 하나 가져오기
# 여기서 모든 건강백과에 있는 병들의 코드르 알아내서 리스트 생성 후 그 리스트에서 이 코드를 반복문으로 실행할 것.
import requests
from bs4 import BeautifulSoup

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# 정보를 저장할 딕셔너리 (div class name = stress)
info_dict = {}

# 각 섹션의 제목에 해당하는 h3 태그와 그에 따른 설명 p 태그를 찾아 딕셔너리에 저장
for h3_tag in soup.find_all("h3", class_="stress"):
    title = h3_tag.get_text().strip()
    content = h3_tag.find_next_sibling("p", class_="txt").get_text().strip()
    info_dict[title] = content
    
#tmp_profile안에 진료과 정보 있음. (진료과도 포함시키기 위함)
tmp_profile_text = soup.find('div', class_='tmp_profile').get_text(strip=True)
info_dict['진료과'] = tmp_profile_text

print(info_dict)
```

```python
# TF-IDF 

import os
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# "sick" 폴더 안의 모든 CSV 파일 경로 가져오기
directory = "sick"
csv_files = [file for file in os.listdir(directory) if file.endswith(".csv")]
documents = {}

# 각 병명에 해당하는 증상들을 저장할 딕셔너리 초기화
for file in csv_files:
    file_path = os.path.join(directory, file)
    disease_name = os.path.splitext(file)[0]
    documents[disease_name] = []

# 각 CSV 파일을 읽어서 병명과 증상을 딕셔너리에 저장
for file in csv_files:
    file_path = os.path.join(directory, file)
    disease_name = os.path.splitext(file)[0]
    df = pd.read_csv(file_path)
    for column in df.columns:
        documents[disease_name].extend(df[column].astype(str).tolist())

# TF-IDF 벡터화 객체 생성
tfidf_vectorizer = TfidfVectorizer()

# 각 병명별로 TF-IDF 행렬 계산하여 중요한 키워드 추출
for disease, symptoms in documents.items():
    tfidf_matrix = tfidf_vectorizer.fit_transform(symptoms)
    feature_names = tfidf_vectorizer.get_feature_names_out()
    sorted_indices = tfidf_matrix.toarray().argsort(axis=1)[:, ::-1]
    top_keywords = [feature_names[ind] for ind in sorted_indices[:, :10]]
    print(f"병명: {disease}")
    print("중요한 키워드:", top_keywords)
    print()
```
