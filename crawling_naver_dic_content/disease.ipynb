```python
# 건강 백과의 모든 병의 이름부터 가져오기
import requests
from bs4 import BeautifulSoup
import re

page_count = 118
all_names = []

for page_num in range(1, page_count + 1):
    url = f"https://terms.naver.com/list.naver?cid=51007&categoryId=51007&page={page_num}"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # class가 subject안의 title 있음
    subject_elements = soup.find_all(class_='title')
    for subject_element in subject_elements:
        name = subject_element.get_text()
        name = re.sub(r'\[[^\]]+\]', '', name).strip()
        all_names.append(name)

print("수집된 병명:")
for name in all_names:
    print(name)
```

```python
#보다보니 질명의 고유한 id값이 있는 걸 발견핵서 이름으로 반복문을 돌리지 않고 질병 id값을 이용해 리스트를 만들어
#반복문을 돌릴 생각임

import requests
from bs4 import BeautifulSoup

page_count = 118
all_contents_ids = []  

for page_num in range(1, page_count + 1):
    url = f"https://terms.naver.com/list.naver?cid=51007&categoryId=51007&page={page_num}"
    response = requests.get(url)
    
    soup = BeautifulSoup(response.text, 'html.parser')

    contents_id_elements = soup.find_all(class_="__TLIST_PUT")

    # data-contentsid 값을 추출함!
    contents_ids = [element.get("data-contentsid") for element in contents_id_elements]
    all_contents_ids.extend(contents_ids)
    
print(all_contents_ids)
```

```python
import os
import pandas as pd
from tqdm import tqdm

# 폴더 경로 설정
folder_name = "sick"
if not os.path.exists(folder_name):
    os.makedirs(folder_name)

# 전체 데이터를 저장할 딕셔너리 초기화
all_info_dict = {}

# 기본 URL 설정
base_url = "https://terms.naver.com/entry.naver?docId="

# tqdm을 사용하여 반복문 실행
for content_id in tqdm(all_contents_ids, desc="진행 상황"):
    # URL 설정
    url = base_url + content_id

    # 페이지 요청
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # 정보를 저장할 딕셔너리 초기화
    info_dict = {}

    # 섹션 제목과 내용을 딕셔너리에 저장
    for h3_tag in soup.find_all("h3", class_="stress"):
        title = h3_tag.get_text().strip()
        content = h3_tag.find_next_sibling("p", class_="txt").get_text().strip()
        info_dict[title] = content

    # 진료과 정보를 추출하여 딕셔너리에 저장
    tmp_profile_text = soup.find('div', class_='tmp_profile').get_text(strip=True)
    info_dict['진료과'] = tmp_profile_text

    # 해당 문서의 정보를 전체 딕셔너리에 추가
    all_info_dict[content_id] = info_dict

    # 데이터프레임 생성
    df = pd.DataFrame.from_dict(info_dict, orient='index', columns=['내용'])

    # 파일명 설정
    file_name = f"{content_id}.csv"

    # 파일 경로 설정
    file_path = os.path.join(folder_name, file_name)

    # CSV 파일로 저장
    df.to_csv(file_path, encoding='utf-8-sig')

print("CSV 파일이 저장된 폴더:", folder_name)
```

```python
# 리스트를 조회하면서 병의 정보를 가져오는 코드 완성본 병명.csv로 저장이 됩니다. 폴더명은 알아서 바꾸셔도 됩니다.

import os
import pandas as pd
from bs4 import BeautifulSoup
import requests
from tqdm import tqdm

# 여기 폴더 이름 바꾸셔도 됩니다.
folder_name = "sick"
if not os.path.exists(folder_name):
    os.makedirs(folder_name)

base_url = "https://terms.naver.com/entry.naver?docId="

# tqdm을 사용하여 진행상황 확인 할 수 있게 함.
for content_id in tqdm(all_contents_ids, desc="진행 상황"):
    try:
        url = base_url + content_id

        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        disease_name = soup.find("h2", class_="headword").get_text().strip()
        # 모든 병명의 공백을 제거하기 위해서 넣었습니다. 뺴도 됩니다.
        disease_name = disease_name.replace(" ", "")
        info_dict = {"병명": disease_name}

        for h3_tag in soup.find_all("h3", class_="stress"):
            title = h3_tag.get_text().strip()
            content = h3_tag.find_next_sibling("p", class_="txt").get_text().strip()
            info_dict[title] = content
        
        # 진료과 추가함
        tmp_profile_text = soup.find('div', class_='tmp_profile').get_text(strip=True)
        info_dict['진료과'] = tmp_profile_text
        df = pd.DataFrame.from_dict(info_dict, orient='index', columns=[content_id])
        df = df.transpose()

        # 파일명 설정
        file_name = f"{disease_name}.csv"

        # 파일 경로 설정
        file_path = os.path.join(folder_name, file_name)

        # CSV 파일로 저장
        df.to_csv(file_path, encoding='utf-8-sig', index=False)

    except Exception as e:
        print(f"Error occurred for content_id: {content_id}, {e}")

print("CSV 파일이 저장된 폴더:", folder_name)
```
