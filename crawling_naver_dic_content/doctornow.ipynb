```python
# 닥터나우의 실시간 상담 내용 가져오기
import requests
from bs4 import BeautifulSoup

url = "https://doctornow.co.kr/content/qna/realtime"

# URL에서 HTML 가져오기
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# 질문 태그 선택 및 추출
question_tags = soup.find_all("h2", class_="sc-c2facd57-5 gHXotb")

# 질문과 관련된 정보 출력
for tag in question_tags:
    question = tag.text.strip()
    print(question)
```

```python
# 첫페이지에서 닥터나우 실시간 questionid값 가져오기
import requests
import json
from datetime import datetime, timedelta

# 현재 시간
now = datetime.now()

# 24시간 이전의 시간
twenty_four_hours_ago = now - timedelta(hours=24)

# 웹 페이지의 URL 설정
url = "https://doctornow.co.kr/content/qna"
response = requests.get(url)
html_text = response.text

# 문자열에서 JSON 데이터를 추출
json_data_start = html_text.find('{"props":')
json_data_end = html_text.find('</script>', json_data_start)
json_str = html_text[json_data_start:json_data_end]

# JSON 데이터 파싱
json_data = json.loads(json_str)

# 24시간 이내의 질문들만 추출
questions_within_24_hours = []

# 모든 질문에 대해 반복
for element in json_data['props']['pageProps']['dehydratedState']['queries'][0]['state']['data']['pages'][0]['newestQuestionCardCursorPage']['elements']:
    question_time = datetime.fromtimestamp(element['question']['createdAt'] / 1000)
    if question_time > twenty_four_hours_ago:
        questions_within_24_hours.append(element['question']['questionPid'])

# 결과 출력
print("24시간 이내의 질문의 questionPid:")
for question_pid in questions_within_24_hours:
    print(question_pid)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Selenium 웹 드라이버 설정 (Chrome 드라이버 사용)
driver = webdriver.Chrome()

# 페이지 열기
driver.get("https://doctornow.co.kr/content/qna")

# 초기 스크롤
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

# 스크롤 후 기다릴 시간
scroll_pause_time = 2

# 스크롤을 더 이상 할 수 없을 때까지 반복
while True:
    # 현재 페이지의 질문 수
    num_questions = len(driver.find_elements(By.CLASS_NAME, "QuestionCard__Container-sc-1lqdrpr-0"))

    # 페이지 스크롤
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # 일정 시간 대기
    time.sleep(scroll_pause_time)

    # 새로운 질문이 추가되었는지 확인
    new_num_questions = len(driver.find_elements(By.CLASS_NAME, "QuestionCard__Container-sc-1lqdrpr-0"))
    if new_num_questions == num_questions:
        break

# 모든 질문의 questionPid 추출
question_pids = []
question_elements = driver.find_elements(By.CLASS_NAME, "QuestionCard__Container-sc-1lqdrpr-0")
for question_element in question_elements:
    question_pid = question_element.get_attribute("data-pid")
    question_pids.append(question_pid)

# 결과 출력
print("전체 질문의 questionPid:")
for question_pid in question_pids:
    print(question_pid)

# 드라이버 종료
driver.quit()
```
